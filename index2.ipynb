```python
#Removing Outliers
#Creating a list of columns that have outliers
columns_with_outliers = [' bank_asset_value', ' residential_assets_value', ' commercial_assets_value' ]

#Creating a loop for removing outliers
for col in columns_with_outliers:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)

   # Calculate IQR
    iqr = q3 - q1

  # using a dynamic multiplier instead of using a fixed multiplier eg 1.5
    lower_bound = q1 - (1.5 * iqr * (len(df[col])/(len(df[col]) + 1)))
    upper_bound = q3 + (1.5 * iqr * (len(df[col])/(len(df[col]) + 1)))
    df[col] = df[col][~((df[col] < lower_bound) | (df[col] > upper_bound))]

#Verifying for changes 
df.info()
```

```python
def missing_values(df):
    # checks sum of missing values in the data
    missing = df.isnull().sum()
    # Create a percentage of missing values in data.
    percentage = (df.isnull().sum())/len(df)
    # Store in a dataframe
    miss = pd.DataFrame({"missing values":missing, "Percentage of missing values": percentage})
    

    return miss 

Missing_Values = missing_values(df)
print (Missing_Values)
```

## kNN Model


We built and evaluated a kNN model using iterative hyperparameter tuning. 

Before training our kNN model, we performed feature scaling on the numerical features. This step is crucial for kNN since the algorithm relies on distance calculations. Scaling ensures that our features have similar scales and ranges, leading to accurate comparisons.

We fitted the scaler on our training data to capture the mean and standard deviation of each feature. Then, we used the same scaler to transform both the training and testing data. This ensures that the data in both sets has a mean of 0 and a standard deviation of 1.

The core of our model lies in the iterative process of hyperparameter tuning for our kNN model. Our goal here is to find the optimal value of `k`, which determines the number of neighbors to consider during predictions.

We implemented a loop that iterates through different values of `k` (usually from 1 to a predefined upper limit, like 20). For each `k`, we created and trained a kNN model using our scaled training data. We then made predictions on our scaled testing data and evaluated the accuracy using the `accuracy_score` function from `sklearn.metrics`.

Throughout this process, we kept track of the best accuracy achieved and the corresponding `k` value. This information will guide us in selecting the final `k` for our best-performing model.

Once our loop finishes, we have the optimal `k` value. We then trained our final kNN model using the entire scaled training dataset and the chosen `k`. This model is now ready to make predictions on new, unseen data.

With our final kNN model, we generated predictions on the scaled testing data. To assess its performance, we used various evaluation metrics including accuracy, precision, recall, a confusion matrix, and a classification report. These metrics provide insights into the model's ability to classify instances accurately and its overall performance.


```python
# Iterative hyperparameter tuning
best_accuracy = 0
best_k = None

for k in range(1, 21):  # Try different values of k
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_scaled, y_train)
    y_pred = knn_model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

# Training the best kNN model on the full training set
final_knn_model = KNeighborsClassifier(n_neighbors=best_k)
final_knn_model.fit(X_train_scaled, y_train)

# Making predictions on the test set
y_pred_final = final_knn_model.predict(X_test_scaled)

# Evaluating the final model
final_accuracy = accuracy_score(y_test, y_pred_final)
precision = precision_score(y_test, y_pred_final, average='weighted')
recall = recall_score(y_test, y_pred_final, average='weighted')
conf_matrix_final = confusion_matrix(y_test, y_pred_final)
classification_rep_final = classification_report(y_test, y_pred_final)

print("Best k:", best_k)
print(f"Best Accuracy: {best_accuracy:.2f}")
print("Final Accuracy:", final_accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("Confusion Matrix:")
print(conf_matrix_final)
print("Classification Report:")
print(classification_rep_final)
```

## GAUSSIAN NAIVE BAYES CLASSIFICATION


For this classification we shall look at multiple observations and conditional probabilities to help increase predictions of loan approval and make the right decisions.We shall do this by implementing point estimates for the conditional probabilities of a feature for a given class, get the prosterior class probabilities of the given observation across all classes, by picking the class with the largest numerator, and finally determine the overall accuracy of our recently created Gaussian Naive Bayes classifier, you'll need to generate predictions for all of the rows in the dataset

```python
# Calculating the mean and standard deviation for each feature within each class for the training set
# to serve as our a priori distribution estimate to determine the posterior likelihood of an observation belonging to one class versus the other.
train = pd.concat([X_train, y_train], axis=1)
aggs = df.groupby(' loan_status').agg(['mean', 'std'])
aggs
```

```python
#Defining a function to calculate the point estimate for the conditional probability of a feature value for a given class
from scipy import stats
def p_x_given_class(obs_row, feature, class_):
    mu = aggs[feature]['mean'][class_]
    std = aggs[feature]['std'][class_]
    
    # Observation
    obs = obs_row[feature] 
    
    p_x_given_y = stats.norm.pdf(obs, loc=mu, scale=std)
    return p_x_given_y
p_x_given_class(X_train.iloc[0], X.columns[0], 0)
```

```python

#Defining a prediction function
def predict_class(obs_row):
    c_probs = []
    for c in range(2):
        # Initialize probability to relative probability of class
        p = len(y_train[y_train == c])/len(y_train) 
        for feature in X.columns:
            p *= p_x_given_class(obs_row, feature, c)
        c_probs.append(p)
    return np.argmax(c_probs)

# we are going to apply the above function to the train and test set

y_hat_train = [predict_class(X_train.iloc[idx]) for idx in range(len(X_train))]
y_hat_test = [predict_class(X_test.iloc[idx]) for idx in range(len(X_test))]

# we are going to calculate the training and test set accuracy
residuals_train = y_hat_train == y_train
acc_train = residuals_train.sum()/len(residuals_train)

residuals_test = y_hat_test == y_test
acc_test = residuals_test.sum()/len(residuals_test)
print('Training Accuracy: {}\tTesting Accuracy: {}'.format(acc_train, acc_test))
```
