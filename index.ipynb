```python
#Checking for outliers
#Creating a list of columns to check for outliers
outlier_checklist = [' income_annum', ' loan_amount', ' loan_term', ' cibil_score', ' residential_assets_value', ' commercial_assets_value', ' luxury_assets_value', ' bank_asset_value']

# Creating a loop for creating a boxplot for each column
for col in outlier_checklist:
    sns.boxplot(x = df[col])
    plt.show()
```

```python
#Removing Outliers
#Creating a list of columns that have outliers
columns_with_outliers = [' bank_asset_value', ' residential_assets_value', ' commercial_assets_value' ]

#Creating a loop for removing outliers
for col in columns_with_outliers:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)

   # Calculate IQR
    iqr = q3 - q1

  # using a dynamic multiplier instead of using a fixed multiplier eg 1.5
    lower_bound = q1 - (1.5 * iqr * (len(df[col])/(len(df[col]) + 1)))
    upper_bound = q3 + (1.5 * iqr * (len(df[col])/(len(df[col]) + 1)))
    df[col] = df[col][~((df[col] < lower_bound) | (df[col] > upper_bound))]

#Verifying for changes 
df.info()
```

```python
def missing_values(df):
    # checks sum of missing values in the data
    missing = df.isnull().sum()
    # Create a percentage of missing values in data.
    percentage = (df.isnull().sum())/len(df)
    # Store in a dataframe
    miss = pd.DataFrame({"missing values":missing, "Percentage of missing values": percentage})
    

    return miss 

Missing_Values = missing_values(df)
print (Missing_Values)
```

```python
#Droping Missing Values
df.dropna(inplace=True)

#Verifying changes 
df.info()
```

```python
#Label encoding to represent categorical variables 
#For the ' loan_status' column
df[' loan_status']=df[' loan_status'].map(lambda x: 1 if x == ' Approved' else 0)

#For the ' self_employed' column
df[' self_employed']=df[' self_employed'].map(lambda x: 1 if x == ' Yes' else 0)

#For the ' education' column
df[' education']=df[' education'].map(lambda x: 1 if x == ' Graduate' else 0)

#Checking for Changes
df.head()
```

# Modelling

With a thorough understanding of the dataset gained through Exploratory Data Analysis (EDA), we are now ready to move on to the model building phase. Our objective is to develop a predictive model that can accurately predict loan approval outcomes based on the provided features.


```python
#Preparing data for model creation

#Setting Up Variables
Y = df[' loan_status']
X = df.drop(' loan_status', axis = 1)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Scaling the numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## i) K-Nearest Neighbors (KNN) as the Base Model

In the early stages of model development, we've opted for the K-Nearest Neighbors (KNN) algorithm as our foundational model. This choice is rooted in its simplicity, interpretability, and alignment with our project's objectives.

### Why KNN as the Base Model?

- **Simplicity and Intuition:** KNN's straightforward approach, making predictions based on nearby data points, is easily comprehensible and interpretable.

- **Data-Driven Decisions:** KNN's predictions directly reflect data patterns, fitting our data-centric decision-making philosophy.

- **Non-Parametric Nature:** KNN's adaptability to various data distributions without making assumptions suits diverse scenarios.

- **Baseline Performance:** By adopting KNN as our starting point, we establish a performance baseline for evaluating advanced models.

### Next Steps in KNN Model Development

Before training, we standardized numerical features to ensure precise distance calculations, a fundamental element of kNN. Scaling enhances the accuracy of comparing feature values.

The core of our model lies in hyperparameter tuning for kNN. We performed iterative training and evaluation with different k values (typically 1 to 20), tracking optimal accuracy to guide k selection.

Following the tuning process, we identified the best k value. The final kNN model was then trained on the entire scaled dataset, poised for predictions.

Using this model, we made predictions on scaled test data. Performance evaluation encompassed accuracy, precision, recall, confusion matrix, and classification report. These metrics provided insights into classification accuracy and overall effectiveness of the model.



```python
# Iterative hyperparameter tuning
best_accuracy = 0
best_k = None

for k in range(1, 21):  # Try different values of k
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_scaled, y_train)
    y_pred = knn_model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

# Training the best kNN model on the full training set
final_knn_model = KNeighborsClassifier(n_neighbors=best_k)
final_knn_model.fit(X_train_scaled, y_train)

# Making predictions on the test set
y_pred_final = final_knn_model.predict(X_test_scaled)

# Evaluating the final model
final_accuracy = accuracy_score(y_test, y_pred_final)
precision = precision_score(y_test, y_pred_final, average='weighted')
recall = recall_score(y_test, y_pred_final, average='weighted')
conf_matrix_final = confusion_matrix(y_test, y_pred_final)
classification_rep_final = classification_report(y_test, y_pred_final)
```

In this section, we showcase the evaluation metrics derived from the first code set, including the F1 score, using a structured dictionary and a DataFrame for clarity. Additionally, we visually represent the model's performance through a well-organized confusion matrix, enhancing our understanding of its predictive abilities across different classes. This combined approach provides a comprehensive insight into the kNN model's effectiveness and aids in informed decision-making.

```python
# Calculate additional metrics
f1_final = f1_score(y_test, y_pred_final, average='weighted')

# Create a dictionary to store evaluation metrics
metrics_dict = {
    'accuracy': [best_accuracy, final_accuracy],
    'precision': [precision, precision_score(y_test, y_pred_final, average='weighted')],
    'recall': [recall, recall_score(y_test, y_pred_final, average='weighted')],
    'f1_score': [f1_final, f1_score(y_test, y_pred_final, average='weighted')]
}

# Create a DataFrame for visualization
results_df = pd.DataFrame(metrics_dict, index=['train', 'test'])

# Print the DataFrame
print(results_df)

# Compute confusion matrix
conf_matrix_final = confusion_matrix(y_test, y_pred_final)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(conf_matrix_final)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

KNN model demonstrated consistent and reliable performance across both the training and test datasets. With accuracy, precision, recall, and F1 score all hovering around 0.92, the model showcased its ability to make balanced predictions. The compact confusion matrix further reinforced its reliability, with only 22 misclassifications out of 315 instances in one class and 46 misclassifications out of 521 instances in another. These results underscore the model's robustness and effectiveness in accurate classification tasks.

After observing the kNN model's consistent performance, we opted to transition to a logistic regression model. This decision is driven by logistic regression's proficiency in binary classification and its capacity to offer interpretable insights into feature influences. Moreover, given its suitability for linear decision boundaries, we anticipate that logistic regression could align well with the dataset's structure. ## ii) Logistic Regression Model
The goal of building a Logistic Regression model is to predict the probability of a binary outcome (e.g., whether a loan application will be approved or rejected) based on input features. Scikit-learn, a popular machine learning library, provides tools to create, train, and evaluate such models.

```python
# Create a Logistic Regression model
logreg_model = LogisticRegression(random_state=1)

# Train the model on the training data
logreg_model.fit(X_train_scaled, y_train)

# Make predictions on the test data
y_pred = logreg_model.predict(X_test_scaled)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(conf_matrix)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()

# Accessing coefficients and intercept
coefficients = logreg_model.coef_
intercept = logreg_model.intercept_

# Displaying the coefficients and intercept
print("Logistic Regression Model")
print("Accuracy: {:.5f}".format(accuracy))
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nCoefficients:")
for feature, coef in zip(X_train.columns, coefficients[0]):
    print(f"{feature}: {coef:.5f}")
print("\nIntercept:", intercept[0])
```

The logistic regression model achieved an accuracy of 0.91627, accurately classifying instances. Its coefficients highlighted features' impact on predictions. Notably, higher loan_amount, cibil_score, and income_annum increased prediction likelihood. Comparing to the kNN model's 0.92 accuracy and balanced metrics, logistic regression offers interpretability via coefficients, revealing feature influence. While kNN excels in capturing nonlinear patterns, logistic regression's insights can inform decision-making. The choice between models hinges on interpretability needs and data characteristics.

## iii) Decision Trees

Our exploration of predictive models continues with the implementation of the decision tree model. Decision trees offer a structured approach to classification tasks, creating a sequence of binary decisions based on features to arrive at a prediction. This model's simplicity and interpretability make it a valuable addition to our analysis. By branching through feature-space, decision trees capture complex relationships within the data. We anticipate that decision trees will excel in scenarios where interactions between features play a crucial role in classification outcomes. Through this exploration, we aim to uncover the unique strengths of decision trees and their potential to enhance our predictive capabilities.

```python
#creating a fuction to get Evaluation metrics for classification
def results(y_train,y_train_pred,y_test,y_test_preds):
    test_para=[accuracy_score(y_test,y_test_preds),precision_score(y_test,y_test_preds),
               recall_score(y_test,y_test_preds),f1_score(y_test,y_test_preds)]
    train_para=[accuracy_score(y_train,y_train_pred),precision_score(y_train,y_train_pred)
     ,recall_score(y_train,y_train_pred),f1_score(y_train,y_train_pred)]
    para=['accuracy','precision','recall','f1_score']
    dic ={'parameter':para,"train":train_para,'test':test_para}
    parameter_df = pd.DataFrame(dic)
    return parameter_df
```

```python
# Creating a decision tree classifier and fitting it on the training data
# Instantiating the model
clf = DecisionTreeClassifier()
#Fiting the model to the training data
clf.fit(X_train_scaled, y_train)
```

```python
 # make predictions on the train and test data 
y_train_pred=clf.predict(X_train_scaled)
y_test_preds = clf.predict(X_test_scaled)

# viewing our evaluation metrics
results(y_train,y_train_pred,y_test,y_test_preds)
```

```python
#displaying test confusion matrix for our first DecisionTreeClassifier model
ConfusionMatrixDisplay.from_estimator(estimator=clf, X=X_test_scaled, y=y_test);
```

```python
#Defining a function to calculate the point estimate for the conditional probability of a feature value for a given class

def p_x_given_class(obs_row, feature, class_):
    mu = aggs[feature]['mean'][class_]
    std = aggs[feature]['std'][class_]
    
    # Observation
    obs = obs_row[feature] 
    
    p_x_given_y = stats.norm.pdf(obs, loc=mu, scale=std)
    return p_x_given_y
p_x_given_class(X_train.iloc[0], X.columns[0], 0)
```

```python
#Defining a prediction function
def predict_class(obs_row):
    c_probs = []
    for c in range(2):
        # Initialize probability to relative probability of class
        p = len(y_train[y_train == c])/len(y_train) 
        for feature in X.columns:
            p *= p_x_given_class(obs_row, feature, c)
        c_probs.append(p)
    return np.argmax(c_probs)

# we are going to apply the above function to the train and test set

y_hat_train = [predict_class(X_train.iloc[idx]) for idx in range(len(X_train))]
y_hat_test = [predict_class(X_test.iloc[idx]) for idx in range(len(X_test))]

# we are going to calculate the training and test set accuracy
residuals_train = y_hat_train == y_train
acc_train = residuals_train.sum()/len(residuals_train)

residuals_test = y_hat_test == y_test
acc_test = residuals_test.sum()/len(residuals_test)
print('Training Accuracy: {}\tTesting Accuracy: {}'.format(acc_train, acc_test))
```
